```python
RACES = {
    "HUMAN": 1,
	"ORC": 2,
	"DWARF": 3,
	"NELF": 4,
	"UNDEAD": 5,
	"TAUREN": 6,
	"GNOME": 7,
	"TROLL": 8,
	"GOBLIN": 9,
	"BELF": 10,
	"DRAENEI": 11,
	"WORGEN": 22,
	"PANDAREN": 24,
	"NIGHTBORNE": 27,
	"HIGHMOUNTAIN_TAUREN": 28,
	"VOID_ELF": 29,
	"LIGHTFORGED_DRAENEI": 30,
	"ZANDALARI_TROLL": 31,
	"KUL_TIRAN": 32,
	"DARK_IRON_DWARF": 34,
	"VULPERA": 35,
	"MAGHAR_ORC": 36,
	"MECHAGNOME": 37
}

GENDER_IDS = [0, 1]
with open("Customizations.h", "w") as myfile:           
    myfile.write("#pragma once\n\n")
    myfile.write("#include <string>\n")
    myfile.write("#include <map>\n")
    myfile.write("#include <unordered_map>\n")
    myfile.write("#include <array>\n\n")
    myfile.write("enum class RaceIDs;\n\n")
    myfile.write("namespace LiMorph {\n")
    myfile.write("namespace Customizations {\n\n")
    for race_name, race in RACES.items():
        for gender in GENDER_IDS:
            if gender:
                ge = "F"
            else:
                ge = "M"
            ChrModelID = chrracexchrmodel.loc[chrracexchrmodel["ChrRacesID"] == race]["ChrModelID"].values.tolist()
            ChrModelID = ChrModelID[1] if gender else ChrModelID[0]
            chrcustomizationoption_sorted = chrcustomizationoption[chrcustomizationoption["ChrModelID"] == ChrModelID].sort_values("ID")
            options = chrcustomizationoption_sorted["Name_lang"].values.tolist()
            options = [s.replace(" ", "").lower() for s in options]


            myfile.write("const uint8_t " + race_name + "_" + ge + "_N_CUSTOMIZATIONS = " + str(len(options)) + ";\n");
            
            myfile.write("const std::unordered_map<std::string, int> " + race_name + "_" + ge + "_CUSTOMIZATIONS {\n");
           # myfile.write("const char* ")
            option_order_index = 0
            for o in options:
                myfile.write('\t{"' + o + '", ' + str(option_order_index) + '},' + "\n")
                option_order_index += 1
            myfile.write("};\n\n")
            

            ChrCustomizationOptionID = chrcustomizationoption_sorted[chrcustomizationoption["ChrModelID"] == ChrModelID][["ID"]]
            ChrCustomizationOptionID = ChrCustomizationOptionID.values[:,0].tolist()
            choices = chrcustomizationchoice[chrcustomizationchoice["ChrCustomizationOptionID"].isin(ChrCustomizationOptionID)]
            
            max_choices = 0
            for i in ChrCustomizationOptionID:
                choices_i = choices[choices["ChrCustomizationOptionID"] == i].sort_values("OrderIndex")["ID"].values.tolist()
                if len(choices_i) > max_choices:
                    max_choices = len(choices_i)
            
            myfile.write("const std::array<uint32_t, 64> " + race_name + "_" + ge + "_NATIVE_CUSTOMIZATIONS {\n");
            for i in ChrCustomizationOptionID:
                choices_i = choices[choices["ChrCustomizationOptionID"] == i].sort_values("OrderIndex")["ID"].values.tolist()
                myfile.write("\t" + str(i) + ",\n")
                myfile.write("\t" + str(choices_i[0]) + ",\n")
            myfile.write("};\n\n")
            
            myfile.write("const std::array<int, " + str(len(options)) + "> " + race_name + "_" + ge + "_CUSTOMIZATION_N_CHOICES {");
            for i in ChrCustomizationOptionID:
                choices_i = choices[choices["ChrCustomizationOptionID"] == i].sort_values("OrderIndex")["ID"].values.tolist()
                myfile.write(str(len(choices_i)) + ', ')
            myfile.write("};\n\n")
            
            myfile.write("const std::array<uint32_t, 32> " + race_name + "_" + ge + "_CUSTOMIZATION_OPTION_IDS {");
            for i in ChrCustomizationOptionID:
                myfile.write(str(i) + ', ')
            myfile.write("};\n\n")
            
            myfile.write("const std::array<std::array<int," + str(max_choices) + ">, " + str(len(options)) + "> " + race_name + "_" + ge + "_CUSTOMIZATION_CHOICES {{\n");
            for i in ChrCustomizationOptionID:
                choices_i = choices[choices["ChrCustomizationOptionID"] == i].sort_values("OrderIndex")["ID"].values.tolist()
                myfile.write("\t{{")
                for k in choices_i:
                    myfile.write(str(k) + ', ')
                myfile.write("}},\n")
            myfile.write("}};\n\n")

    myfile.write("\n\n")
    myfile.write("const std::array<uint32_t, 32>& getCustomizationOptionIDs(int race_id, int gender) {\n")
    myfile.write("switch(static_cast<RaceIDs>(race_id)) {\n");
    for race_name, race in RACES.items():
        myfile.write("\tcase RaceIDs::" + race_name + ":\n");

        myfile.write("\tif(gender) \n");
        myfile.write("\t\treturn " + "Customizations::" + race_name + "_F_CUSTOMIZATION_OPTION_IDS;\n");
        myfile.write("\telse\n");
        myfile.write("\t\treturn " + "Customizations::" + race_name + "_M_CUSTOMIZATION_OPTION_IDS;\n");

    myfile.write("}\n");
    myfile.write("}\n\n");
        
    myfile.write("const std::unordered_map<std::string, int>& getCustomizations(int race_id, int gender) {\n")
    myfile.write("switch(static_cast<RaceIDs>(race_id)) {\n");
    for race_name, race in RACES.items():
        myfile.write("\tcase RaceIDs::" + race_name + ":\n");

        myfile.write("\tif(gender) \n");
        myfile.write("\t\treturn " + "Customizations::" + race_name + "_F_CUSTOMIZATIONS;\n");
        myfile.write("\telse\n");
        myfile.write("\t\treturn " + "Customizations::" + race_name + "_M_CUSTOMIZATIONS;\n");

    myfile.write("}\n");
    myfile.write("}\n\n");
    
    myfile.write("int getNumberOfCustomizations(int race_id, int gender) {\n")
    myfile.write("switch(static_cast<RaceIDs>(race_id)) {\n");
    for race_name, race in RACES.items():
        myfile.write("\tcase RaceIDs::" + race_name + ":\n");

        myfile.write("\tif(gender) \n");
        myfile.write("\t\treturn " + "Customizations::" + race_name + "_F_N_CUSTOMIZATIONS;\n");
        myfile.write("\telse\n");
        myfile.write("\t\treturn " + "Customizations::" + race_name + "_M_N_CUSTOMIZATIONS;\n");

    myfile.write("}\n");
    myfile.write("}\n\n");
    
    myfile.write("int getNumberOfChoices(int race_id, int gender, int option) {\n")
    myfile.write("switch(static_cast<RaceIDs>(race_id)) {\n");
    for race_name, race in RACES.items():
        myfile.write("\tcase RaceIDs::" + race_name + ":\n");

        myfile.write("\tif(gender) \n");
        myfile.write("\t\treturn " + "Customizations::" + race_name + "_F_CUSTOMIZATION_N_CHOICES[option];\n");
        myfile.write("\telse\n");
        myfile.write("\t\treturn " + "Customizations::" + race_name + "_M_CUSTOMIZATION_N_CHOICES[option];\n");

    myfile.write("}\n");
    myfile.write("}\n\n");
    
    myfile.write("int getChoiceID(int race_id, int gender, int option, int choice) {\n")
    myfile.write("switch(static_cast<RaceIDs>(race_id)) {\n");
    for race_name, race in RACES.items():
        myfile.write("\tcase RaceIDs::" + race_name + ":\n");

        myfile.write("\tif(gender) \n");
        myfile.write("\t\treturn " + "Customizations::" + race_name + "_F_CUSTOMIZATION_CHOICES[option][choice];\n");
        myfile.write("\telse\n");
        myfile.write("\t\treturn " + "Customizations::" + race_name + "_M_CUSTOMIZATION_CHOICES[option][choice];\n");

    myfile.write("}\n");
    myfile.write("}\n\n");
    
    myfile.write("const std::array<uint32_t, 64>& getNativeCustomizations(int race_id, int gender) {\n")
    myfile.write("switch(static_cast<RaceIDs>(race_id)) {\n");
    for race_name, race in RACES.items():
        myfile.write("\tcase RaceIDs::" + race_name + ":\n");

        myfile.write("\tif(gender) \n");
        myfile.write("\t\treturn " + "Customizations::" + race_name + "_F_NATIVE_CUSTOMIZATIONS;\n");
        myfile.write("\telse\n");
        myfile.write("\t\treturn " + "Customizations::" + race_name + "_M_NATIVE_CUSTOMIZATIONS;\n");

    myfile.write("}\n");
    myfile.write("}\n\n");
    
    myfile.write("} // namespace Customizations\n")
    myfile.write("} // namespace LiMorph")
            
```
